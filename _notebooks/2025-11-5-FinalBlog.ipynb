## Night at the Museum (N@tM)
<div class="card pop">
  <p><strong>Short version:</strong> people didn’t just glance; they actually stayed, interacted, and asked thoughtful questions. I’m grateful for that. Teachers, families, and students gave us specific, respectful feedback—and I tried to return that same energy in how I presented.</p>
  <p>Favorite details that landed: small music cues on clicks, smooth transitions between sections, and quiz feedback that nudged (not scolded).
